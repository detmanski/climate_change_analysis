```python
# Extract the next DataFrame
atmospheric_co2_df = pd.read_csv('../Resources/base_data/Atmospheric_CO%E2%82%82_Concentrations.csv')
atmospheric_co2_df.head()
```

```python
#drop unwanted columns
atmospheric_co2_df.drop(columns=["Indicator", "ISO2", "ISO3","CTS_Code", "CTS_Name","CTS_Full_Descriptor", "Source"], axis=1, inplace=True)
atmospheric_co2_df.head()
```

```python
# Remove the 'M' from the 'Date" row 
atmospheric_co2_df["Date"] = atmospheric_co2_df["Date"].map(lambda x: x.replace('M', ''))
atmospheric_co2_df.head()
    
```

```python
# Convert Date column to integer
import datetime as dt 
atmospheric_co2_df.info()
```

```python
atmospheric_co2_df["Date"].astype(int)
```

```python
# Now the Date column can be converted to datetime format
atmospheric_co2_df["Date"] = pd.to_datetime(atmospheric_co2_df["Date"], format='%Y%m')
```

```python
atmospheric_co2_df.rename({"Value": "Amount"}, inplace=True)
```

```python
#Export cleaned data file
atmospheric_co2_df.to_csv('../Resources/cleaned_data/atmospheric_co2_clean.csv', index=False)
```
