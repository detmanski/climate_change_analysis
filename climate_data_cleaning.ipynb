```python
# Extract the next DataFrame
atmospheric_co2_df = pd.read_csv('Resources/Atmospheric_CO%E2%82%82_Concentrations.csv')
atmospheric_co2_df.head()
```

```python
#drop unwanted columns
atmospheric_co2_df.drop(columns=["ObjectId", "Indicator", "ISO2", "ISO3","CTS_Code", "CTS_Name","CTS_Full_Descriptor", "Source"], axis=1, inplace=True)
atmospheric_co2_df.head()
```

```python
# Remove the 'M' from the 'Date" row 
atmospheric_co2_df["Date"] = atmospheric_co2_df["Date"].map(lambda x: x.replace('M', ''))
atmospheric_co2_df.head()
    
```
