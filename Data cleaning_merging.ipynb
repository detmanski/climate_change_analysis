```python
forest_carbon= pd.read_csv("Resources/forest_and_carbon_clean.csv")
forest_carbon=forest_carbon.drop(['ObjectId'], axis=1)
forest_carbon.head(5)
```

```python
forest_carbon_df=forest_carbon.melt(id_vars=["Country","Indicator",'Unit'], 
        var_name="Year",
        value_name="forest_carbon")
forest_carbon_df
```

```python

forest_carbon_df1=forest_carbon_df.replace(['Russian Federation','China, P.R.: Mainland',"Iran, Islamic Rep. of"],['Russia','China','Iran'])
forest_carbon_df1
```

```python
forest_carbon_df1.loc[forest_carbon_df1['Country'] == 'Russia']
```

```python
forest = forest_carbon_df1.pivot_table('forest_carbon', ['Country',"Year"], 'Indicator')
forest
```

```python

df = pd.DataFrame(forest)
df
forest_df=df.reset_index()
forest_df
```

```python
forest_df_new=forest_df['Year'].str[1:5]
forest_df_new
```

```python
forest_df['Year'] =forest_df.index.to_series().map(forest_df_new)
forest_df
```

```python
forest_df.to_csv('Resources/forest_carbon_clean3.csv', index=False)
```
